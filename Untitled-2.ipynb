__Méthode d'ordre 1__

Pour la fonction sole_euler_explicit, nous avons choisi de rajouter un argument. C'est le n qui est le nombre de point que l'on choisi de prendre cela nous évite d'avoir à modifier la fonction pour tester l'erreur. Cela nous fournit la condition d'arret. 

Nous avons également choisi de programmer une seconde fonction euler qui prend un temps final en compte et un temps initial. Cette modifiction nous permet d'observer la fonction sur des domaines différents. La seconde fonction nous permet de nous éviter de fixer un temps initial à 0. 

```python
def solve_euler_explicit(f, x0, dt, n): 
    x = [x0]
    t = [0]
    for k in range(n):
        x.append(x[-1]+dt*f(x[-1]))
        t.append(t[-1]+dt)
    return t, x


def solve_euler_explicit2(f, x0, dt, t0 = 0, tf = 3): 
    x = [x0]
    t = [t0]
    while t[-1] < tf:
        x.append(x[-1]+dt*f(x[-1]))
        t.append(t[-1]+dt)
    return t, x
```

__Méthode d'ordre 2__

Pour le choix de la méthode d'ordre 2, nous avons choisi la méthode de Runge-Kutta d'ordre 2 aussi appelé Heun. 

On rappelle $ x(t +dt) = x(t)$ $+$ $ \int_{t_{0}}^t f(s,x(s))ds$

On approxime ici l'intégrale par l'aire d'un rectangle de hauteur determinée par la valeur de $f$ en $t + \frac{dt}{2}$

> On a donc $x(t + dt) \sim x(t) + f(t + \frac{dt}{2}, x(t + \frac{dt}{2})) \times
dt$

La valeur en $t + \frac{dt}{2}$ de $x$ est calculée de manière approchée par par un développement de Taylor à l'ordre 1:
> $x(t + dt) \sim x(t) + f($ $t + \frac{dt}{2}$ $,$ $ x(t) + \frac{dt}{2}$ $ \times $ $ f(t,x(t))$ $) $ $\times $ $dt$

```python
def solve_runge2(f,x0,dt,N):
    X=[x0]
    T=[0]
    for i in range (N):
        X.append(X[-1]+dt*f(T[-1]+dt/2,X[-1]+dt/2*f(T[-1],X[-1])))
        T.append(T[-1]+dt)
    return T,X
```

Implémenter un pas de temps adaptatif à ces méthodes permettrait de garder une erreur minimale tout en diminuant le nombre de points à calculer.

# Adaptation du pas de temps

Pour cela, on utilise la méthode d'euler avec un pas de temps adaptatif. Cela permet de minimiser l'erreur tout en calculant le nombre minimum de points. Cela permet par la suite de pouvoir utiliser les listes que la fonction renvoie de façon optimale. 

__Montrer que si f est $C^1$, on a pour schéma d'Euler explicite
$e^{j+1}=\Delta t\frac{f(t_{j+1},x^{j+1})-f(t_j,x^j)}{2}+O(\Delta t_j^3)$__

On note $x_{sol}$ la solution exacte du problème de Cauchy à l'itération j:
>$x_{sol}(t_j)=x^j$

>$x_{sol}'=f(t,x_{sol})$

Sachant que f est $C^1$ alors $x_{sol}$ est $C^2$. D'après la formule de Taylor: 
 $x_{sol}(t_j+1)=x_{sol}(t_j)+f(t_j,x_{sol}(t_j))*dt+\frac{1}{2}*\frac{df(t_j,x_{sol}(t_j)}{dt}*dt^2+O(dt^3)$

Or, d'après Euler, $x^{j+1}=x^j+f(t_j,x^j)*dt$
>D'ou, $e^{j+1}=x_{sol}(t_{j+1})-x^{j+1}=\frac{1}{2}*\frac{df(t_j,x^j)}{dt}*dt^2+O(dt^3)$

Or, d'après Taylor, $f(t_{j+1},x^{j+1})=f(t_j,x^j)+\frac{\delta f(t_j,x^j)}{\delta t}*dt+\frac{\delta f(t_j,x^j)}{\delta x}*dx+O(dt^2)+O(dx^2)$

et $(\frac{\delta f(t_j,x^j)}{\delta t}*dt+\frac{\delta f(t_j,x^j)}{\delta x}*dx)*\frac{1}{dt}=\frac{\delta f(t_j,x^j)}{\delta t}+f(t_j,x^j)*\frac{\delta f(t_j,x^j)}{\delta x}=\frac{df(t_j,x^j)}{dt}$


On a donc bien:
>$e^{j+1}=\Delta t\frac{f(t_{j+1},x^{j+1})-f(t_j,x^j)}{2}+O(\Delta t_j^3)$


__Montrer que $e^{j+1}=O(\Delta t^2)$ et que $\Delta t_{new}=\Delta t *\sqrt \frac {Tol_{abs}}{e^{j+1}}$__

On note $E=[t_j,t_{j+1}]*[x^j,x^{j+1}]$

E est un fermé compact de $R^2$ et f est $C^1$ sur E donc la différentielle de f est continue sur E donc bornée sur E

D'ou, $\exists M\in R, \forall (t,x)\in E,\forall (t',x')\in E,|f(t,x)-f(t',x')|\leq M*||(t,x)-(t',x')||$

Or $||(t,x)-(t',x')||$=$\sqrt{\Delta t^2+\Delta x^2}$=$O(\Delta t)$  (car $\Delta x=O(\Delta t)$ selon Taylor appliqué à x à l'ordre 1)

Donc:
>$e^{j+1}=O(\Delta t^2)$

On veut toujours que $e^{j+1}\leq Tol_{abs}$

Si $e^{j+1}=k*Tol_{abs}$ avec $k\in R$ alors il faut diviser l'erreur par k 

Or $e^{j+1}=O(\Delta t^2)$ donc il faut diviser le pas de temps par $\sqrt{k}$

D'ou 
>$\Delta t_{new}=\frac{\Delta t}{\sqrt k}=\Delta t *\sqrt \frac {Tol_{abs}}{e^{j+1}}$


```python
def solve_ivp_euler_explicit_variable_step(f, t0, x0, t_f, dtmin = 1e-16, dtmax = 0.01, atol = 1e-6):
    dt = dtmax/10; # initial integration step
    ts, xs = [t0], [x0]  # storage variables
    t = t0
    ti = 0  # internal time keeping track of time since latest storage point : must remain below dtmax
    x = x0
    dts = [dt]
    while ts[-1] < t_f:
        while ti < dtmax:
            t_next, ti_next, x_next = t + dt, ti + dt, x + dt * f(x)
            x_back = x_next - dt * f(x_next)
            ratio_abs_error = atol / (linalg.norm(x_back-x)/2)
            dt = 0.9 * dt * sqrt(ratio_abs_error)
            if dt < dtmin:
                raise ValueError("Time step below minimum")
            elif dt > dtmax/2:
                dt = dtmax/2
            t, ti, x = t_next, ti_next, x_next
        dt2DT = dtmax - ti # time left to dtmax
        t_next, ti_next, x_next = t + dt2DT, 0, x + dt2DT * f(x)
        ts = vstack([ts,t_next])
        xs = vstack([xs,x_next])
        dts.append(dt)
        t, ti, x = t_next, ti_next, x_next
    return ts, xs, dts


a,b,c = solve_ivp_euler_explicit_variable_step(f1, 0, 1, 3)
plt.plot(a,b)
c = np.array(c)
c = c*15000
plt.plot(a,c)
```

Nous nous intéressons donc à ce que réalise cette fonction. 
Elle prend en argument la fonction que l'on applique à $x$, la condition initiale $x(t_0) = x_0$, le temps final, les pas de temps maximum et minimum ainsi que la tolérence. 

Par la suite, on initialise toutes les variables. Puis, on entre dans la boucle while principale, celle qui nous permet d'obtenir la courbe entre $t_0$ et $t_f$ de la solution de notre équation. 

Dans cette boucle, on a une seconde boucle while. C'est cette seconde boucle qui permet l'adaptation du pas de temps. 
>`t_next, ti_next, x_next = t + dt, ti + dt, x + dt * f(x)`

Cette ligne permet de stocker des valeur temporaires nous permettant d'adapter le pas de temps.
>`x_back = x_next - dt * f(x_next)
ratio_abs_error = atol / (linalg.norm(x_back-x)/2)
dt = 0.9 * dt * sqrt(ratio_abs_error)`

Ce bloc nous permet calculer le nouveau dt adapté en augmentant les marge d'erreur (on divise l'erreur absolue par 2 et on ajoute un coefficient 0,9). 
>`if dt < dtmin:
      raise ValueError("Time step below minimum")
 elif dt > dtmax/2:
      dt = dtmax/2
 t, ti, x = t_next, ti_next, x_next`

Ici, on arrête la résolution si on est inférieur à dtmin, sinon on affecte une nouvelle valeur à dt. Puis , on affecte les nouvelles valeurs aux variable.
>`dt2DT = dtmax - ti # time left to dtmax
 t_next, ti_next, x_next = t + dt2DT, 0, x + dt2DT * f(x)`

A la sortie de la boucle, on adapte donc le pas de temps en soustrayant la somme des dt au dtmax. Cela nous donne le nouveau pas de temps. 

Une fois le pas de temps adapté, on prend le point correspondant on ajoute le temps et la position respectivement aux abscisses et aux ordonnées. Puis on réinitialise ti (qui est l'incrément de la boucle while secondaire) et x et t. La fonction renvoie normalement 2 listes, les positions et les temps. Nous avons ajouter une liste qui comporte tous les dt adaptés ce qui nous permet de visualiser la courbe en orange. Cette courbe repésente l'évolution des dt (multiplié par 15000 de façon à les voir à la bonne échelle) en fonction du temps. On peut voir que le pas de temps diminue en fonction de l'augmentation de la pente de la courbe. 

On s'est intéressé par la suite à l'erreur que donne cette méthode par raport à la méthode avec un pas de temps fixe. On peut observer une erreur environ 2 fois inférieure pour cette méthode. De plus, le temps d'execution est très court pour les 2 méthodes. Même s'il est un peu plus court d'executer le pas de temps fixe. Par la suite on obtient donc le nombre minimum de point pour être précis avec le pas adaptatif.

```python
def erreurMax3(f, t0, x0, t_f):
    diff = []
    a, b, c = solve_ivp_euler_explicit_variable_step(f, t0, x0, t_f, dtmax = 0.001)
    ordonnee = [np.exp(k) for k in a]
    for i in range(len(b)):
        diff.append(abs(b[i]-ordonnee[i]))
    return max(diff)

print(erreurMax1(f1,1, 3))
print(erreurMax3(f1, 0 ,1 ,1))
```

En conclusion, on peut dire que pour des fonctions simple la solution sans adaptation du pas de temps convient. Cependant, lorsque les solutions commencent à être plus complexes ils est plus confortable de travailler sur un pas de temps adaptatif. De plus, l'adaptation du pas de temps permet de minimiser les données tout en ayant une erreur trè faible et donc le traitement est plus simple par la suite.
